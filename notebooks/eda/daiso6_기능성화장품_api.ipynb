```python
import os
import re
import requests
import pandas as pd
import matplotlib.pyplot as plt
from datetime import datetime, timedelta, timezone
import xml.etree.ElementTree as ET
import time
from pathlib import Path

PROJECT_ROOT = Path(r"G:/Final_proj/Total_clear/데이터")
EDA_OUTPUT_DIR = PROJECT_ROOT / "1.EDA_outputs"
EDA_OUTPUT_DIR.mkdir(exist_ok=True)

print(f"저장 경로: {EDA_OUTPUT_DIR}")
```

```python
# 호출할 API URL 입력
url = 'http://apis.data.go.kr/1471000/FtnltCosmRptPrdlstInfoService/getRptPrdlstInq'

dfs = []
page = 1

while True:
    params = {
        'serviceKey': '7e22cdfd1f3b7c77ca6c3dbfda2a2735146d86adbe40aaf01a500b09801c1d92',
        'pageNo': page,
        'numOfRows': 500,
        'type': 'json'
    }

    r = requests.get(url, params=params)
    data = r.json()

    items = data.get('body', {}).get('items', [])
    if not items:
        break

    dfs.append(pd.DataFrame(items))
    print(f"{page}페이지 수집 완료")

    page += 1
    time.sleep(0.2)

final_df = pd.concat(dfs, ignore_index=True)
print(final_df.shape)
```

```python
final_df['ITEM_NAME'] = final_df['ITEM_NAME'].str.strip()
final_df['ITEM_NAME'] = final_df['ITEM_NAME'].str.replace(' ', '', regex=False)

```

```python
final_df.to_csv(EDA_OUTPUT_DIR / "functional_cosmetics.csv", index=False, encoding="utf-8-sig")
print(f"저장 완료: {EDA_OUTPUT_DIR / 'functional_cosmetics.csv'}")
```
